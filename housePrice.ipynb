```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
warnings.filterwarnings('ignore')
```

dataset link : https://www.kaggle.com/datasets/tusharpaul2001/house-price-prediction

```python
data = pd.read_csv('Delhi.csv')
data.head()
```

count rows

```python
len(data)
```

show information

```python
data.describe()
```

```python
data = data.drop('Location',axis=1)
data
```

```python
data.hist(figsize=(20, 20))
```

```python
plt.figure(figsize=(30,15))
sns.heatmap(data.corr(),annot = True)
```

```python
import numpy as np
```

split data

```python
X = data.iloc[:, 1:].values
y = data['Price'].values
```

show rows information

```python
np.shape(y)
```

normalize data

```python
mn = np.min(X,axis=0)
mx = np.max(X,axis=0)
normal_x = (X-mn)/(mx-mn)
normal_x
```

train , test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(normal_x, y,test_size = 0.3)
```

linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

predict

```python
y_pred = model.predict(x_test)
y_pred
```

score

```python
from sklearn.metrics import mean_squared_error,r2_score
MSE = mean_squared_error(y_test,y_pred)
MSE
```

```python
r2_test = r2_score(y_test,y_pred)
r2_test
```

every feature influence , every parametr weight

```python
c = model.coef_
c
```

```python
c0 = model.intercept_
c0
```

```python
y_pred_train = model.predict(x_train)
```

```python
r2_train = r2_score(y_train,y_pred_train)
r2_train
```

```python
print(f'r2_test : {r2_test} \nr2_train : {r2_train}')
```

poly regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
x_poly_train = poly_reg.fit_transform(x_train)
x_poly_test = poly_reg.fit_transform(x_test)
```

```python
model.fit(x_poly_train,y_train)
```

```python
y_pred_train = model.predict(x_poly_train)
y_pred_train
```

```python
y_pred_test = model.predict(x_poly_test)
y_pred_test
```

```python
mse_train = mean_squared_error(y_train,y_pred_train)
mse_test = mean_squared_error(y_test,y_pred_test)
```

```python
print(f'MSE test : {mse_train} \nMSE train : {mse_test}')
```

```python
r2_train = r2_score(y_train,y_pred_train)
r2_test = r2_score(y_test,y_pred_test)
```

```python
print(f'r2_test : {r2_test} \nr2_train : {r2_train}')
```
